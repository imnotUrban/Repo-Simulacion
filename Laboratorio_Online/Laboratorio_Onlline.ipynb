#### <b>1.i) Defina lo que es semilla y para qué sirve</b>
<span style="font-size: 20px;"><b>Contexto:</b></span>
<p>Sea el módulo/librería <b>random</b> aquel que implementa <b>generadores</b> de números <b>pseudo-aleatorios</b> para varias distribuciones.</p>
<span style="font-size: 18px;"><b>¿qué es un iterador?</b></span>
<p>Técnicamente, en Python, un <b>iterador</b> es un objeto que implementa el protocolo del iterador, que consta de los métodos __iter__(): inicialización ;  y __next__(): siguiente valor</p>
<span style="font-size: 18px;"><b>¿qué es un generador?</b></span>
<p>Un <b>generador</b> es un tipo especial de función que no devuelve un solo valor, sino un objeto <b>iterador</b> con una secuencia de valores.</p>
<span style="font-size: 18px;"><b>ejemplo:</b></span>
<p>Tenemos un <b>generador</b> con una secuencia finita que contiene los siguientes números <b>10, 20, 30, 40</b>, el generador los devuelve utilizando la palabra reservada <b>'yield'</b>, generalmente los datos de un <b>generador</b> son secuencias y el estado en el que se encuentre es guardado. Al avanzar no se puede retroceder.</p>
<br>
<span style="font-size: 22px;"><b>Entonces, ¿qué es una semilla y para qué sirve?</b></span>
<p>Una <b>semilla</b> se utiliza para inicializar la base de valores pseudo-aleatorios, por ejemplo la función <b>random.random()</b> nos genera un núemero <b>[0.0, 1.0)</b> de manera equiprobable y además se podría decir que cada vez toma una <b>seed</b> distinta</p>
<span style="font-size: 20px;"><b>Para ejemplificar el uso de una semilla</b></span>
<ul>
<li>(seed= valor_inicialización_generador) =>  números . . .</li>
<li><b>(seed=?)</b> => 0.94 0.05 0.14 0.95 0.43 . . .</li>
<li><b>(seed=?)</b> => 0.21 0.65 0.06 0.76 0.89 . . .</li>
<li><b>(seed=0)</b> => <b>0.08 0.62 0.23 0.03 0.99 <span style="color:red;">0.71 0.25 0.26 0.56 0.69 . . .</span></b></li>
<li><b>(seed=0)</b> => <b>0.08 0.62 0.23 0.03 0.99 . . .</b></li>
<li><b>si no reseteamos la seed y seguimos iterando</b> => <b><span style="color:red;">0.71 0.25 0.26 0.56 0.69 . . .</span></b></li>
</ul>
<p>De esta manera con una <b>semilla</b> podemos replicar la generación de estos números pseudo-aleatorios, útil para recrear expermientos, ejercicios o formulaciones que requieren el uso de datos pseudo-aleatorios.</p>

#### <b>1.ii) Utilice random.random() para generar números aleatorios.</b>
<p>Ejemplo de plot utilizando <b>random.random()</b>, esto debería ser <b>DIFERENTE</b> cada vez que se ejecute</p>

```python
def plot_data(size = 100, color_first5= 'red', color = 'royalblue',  x_label = 'X' ,y_label = 'Y', title = 'Data pseudo-aleatoria, size: ', seed=None):
    """
    Plot ('o') de data pseudo-aleatoria (x,y) equiprobables en el espacio [0,1)x[0,1)
    parámetros:
        size: Tamaño de la muestra > 5
        color_first5: Color de los primeros 5 puntos
        color: Color de los puntos
        x_label: Etiqueta del eje x
        y_label: Etiqueta del eje y
        title: Titulo del grafico
        seed: Semilla para la generacion de numeros aleatorios (default=0)

    return: x,y -> arr
    """
    # If seed parameter is omitted or None, the current system time is used.
    # documentation: https://docs.python.org/3/library/random.html
    random.seed(seed)

    # using random.random() to generate random numbers (x,y) in [0,1)x[0,1)
    x = [random.random() for x in range(size)]
    y = [random.random() for y in range(size)]
    plt.figure(figsize=(10, 6))
    plt.plot(x[:6], y[:6], 'o', color= color_first5, markersize=5, alpha=0.85)
    plt.plot(x[6:], y[6:], 'o', color= color, markersize=5, alpha=0.85)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title + str(size))
    plt.show();
    return x,y


def first5_elem(x,y):
    """
    Imprime la posición de los primeros 5 puntos
    parámetros:
        x: int:arr - x_size: 5 elements
        y: int:arr - x_size: 5 elements
    """
    for index_list, pos  in enumerate(['Primero', 'Segundo', 'Tercer', 'Cuarto', 'Quinto']):
        print(f"{pos} punto: ({x[index_list]} {y[index_list]}")
```

```python
x_pseudo, y_pseudo = plot_data()
first5_elem(x_pseudo[:6], y_pseudo[:6])
```

#### <b>1.iii) Utilice una semilla con random.seed(semilla), y vuelva a ejecutar random.random()</b>
<p>Ejemplo de plot utilizando <b>random.seed(1234)</b>, esto debería ser <b>IGUAL</b> cada vez que se ejecute.</p>

```python
seed = 1234
random.seed(seed)
x_1234_0,y_1234_0 = plot_data(seed=1234)
first5_elem(x_1234_0[:6], y_1234_0[:6])
next5000values_1234_0 = [random.random() for x in range(5000)]
```

### <b>Respuesta 2:</b>

```python
def rand_plot(size=500, rand_arr=None, color = 'cornflowerblue'):
    """
        Hace un plot de barras, barras ordenadas de menor a mayor, boxplot y histograma
        parámetros:
            size: Tamaño de la muestra
            rand_arr: Arreglo pre-generado de números pseudo-aleatorios
            color: Color de las barras
    """
    
    # Si rand_arr no es None, se usa ese arreglo enviado, sino se genera uno nuevo con random.random()
    if rand_arr is None:
        rand_arr = [random.random() for x in range(size)]

    # fig config size: 1x4
    fig, ax = plt.subplots(1, 4, figsize=(16, 7), dpi= 100)

    # Sección 1: Barras
    ax[0].bar(range(0, size), rand_arr, color=color);
    ax[0].set_title('Barras', fontsize=12)

    # Sección 2: Barras ordenadas
    rand_arr.sort()
    ax[1].bar(range(0, size), rand_arr, color=color);
    ax[1].set_title('Barras Ordenadas', fontsize=12)

    # Sección 3: Boxplot para ver el rango
    ax[2].boxplot(rand_arr, vert=False);
    ax[2].set_title('Boxplot', fontsize=12)

    # Sección 4: Histograma para ver la distribución
    ax[3].hist(rand_arr, bins=20, color=color, alpha=0.85);
    ax[3].set_title('Histograma', fontsize=12);
    
    #show plot
    plt.show();
```

<span style="font-size:22px">- random.<b>randint(a,b)</b></span>
<p>Retorna un número entre los parámetros <b>[a,b]</b>: sí a = -100, b = 100, el valor a retornar está entre esos números Enteros, con -100 y 100 inclusive. 'a' y 'b' no tienen que ser <b>int</b>.</p>

<span style="font-size:22px">- random.<b>randrange(a,b)</b></span>
<p>Retorna un número entero entre <b>[a,b)</b>, es decir, incluye 'a' pero no 'b'. También <b>a < b</b> </p>
<p>La gracia de esta función radica en su <b>stepper</b>, el cuál sería un parámetro 'c' del salto. De esta manera podríamos generar números pseudo-aleatorios tal que cumplan una condición, ejemplo que sean pares o impares dentro de cierto rango.</p>

```python
size = 50 # número de elementos
b = 10 # límite superior
step = 2 # step

rand_pares = [random.randrange(0, b, step) for x in range(size)]
print(f"rand_pares: {rand_pares}")
rand_impares = [random.randrange(1, b, step) for x in range(size)]
print(f"rand_impares: {rand_impares}")
```

<p> Para este caso podemos decir que el ciclo es 18 ya que en todas las semillas posibles (desde 0 hasta 18, después se repite) ocurre que el primer y último elemento (el elemento 0 y 18) se repite, por ende entre 0 y 17 hay secuencias de 18 números</p>

### Desde aquí no lo pide pero es para ver qué pasa si se cambia el módulo

```python
def builder(n):
    class NotSoRandom(object):
        def seed(self, a=3):
            """Seed the world's most mysterious random number generator."""
            self.seedval = a
        def random(self):
            """Look, random numbers!"""
            self.seedval = (self.seedval * 3) % n
            return self.seedval

    _inst = NotSoRandom()

    ## se editó esto para evitar problemas de renombre
    seed_ = _inst.seed
    random_ = _inst.random

    secuencia = []
    secuencias = []
    for i in range(n):
        seed_(i) # intentando con otra semilla
        for j in range (n):
            secuencia.append(random_())
        
        secuencias.append(secuencia)
        secuencia = []

    return secuencias

```

### Visualización de secuencias

```python
#parameters
size = 19
h, w = 2, 9 # h*w = size
secuencias = builder(size)

#static
it = 0
fig, ax = plt.subplots(h, w, figsize=(16, 7), dpi= 100)
fig.tight_layout(pad=1.0)
for i in range(h):
    for j in range(w):
        ax[i][j].bar([0], secuencias[it][0], color='red');
        ax[i][j].bar(range(1, size-1), secuencias[it][1:size-1], color='dodgerblue');
        ax[i][j].bar([size-1], secuencias[it][size-1], color='red');
        ax[i][j].set_title('Semilla: '+str(it), fontsize=12)
        it += 1
```
